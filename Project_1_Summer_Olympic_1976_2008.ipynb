# 3.1:  Which country in this chart demonstrates the largest disparity between male and female medal counts, and what might be the reasons behind it?

```python
m_w_medals = df.groupby(['Country']).agg(
    Gold=('Gold', 'sum'),
    Silver=('Silver', 'sum'),
    Bronze=('Bronze', 'sum'),
).sort_values('Gold',ascending=False).reset_index()[:10]

m_w_medals['Total_Medals'] = m_w_medals['Gold'] + m_w_medals['Silver'] + m_w_medals['Bronze']

display(m_w_medals)
```

```python
plt.figure(figsize=(10, 6))

x = m_w_medals['Country']
x1 = m_w_medals['Gold'].head(10)
x2 = m_w_medals['Silver'].head(10)
x3 = m_w_medals['Bronze'].head(10)
x4 = m_w_medals['Total_Medals'].head(10)

# Plotting
plt.scatter(x, x1, color='blue', label='Gold', marker='o')
plt.scatter(x, x2, color='green', label='Silver', marker='s')
plt.scatter(x, x3, color='red', label='Bronze', marker='^')
plt.scatter(x, x4, color='black', label='Total Medals', marker='*')

plt.xlabel('Country')
plt.ylabel('Medals')
plt.title('Top 10 Country have most medals ')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

 # 3.4:  Which sport demonstrates the closest parity between men's and women's medal counts, and what does this suggest about gender inclusivity in that discipline?

```python
some_info = df.groupby(['Sport']).agg(
    Gold=('Gold', 'sum'),
    Silver=('Silver', 'sum'),
    Bronze=('Bronze', 'sum'),
    Men=('Men', 'sum'),
    Women=('Women', 'sum')
).sort_values('Gold',ascending=False).reset_index()

some_info['Total_Medals'] = some_info['Gold'] + some_info['Silver'] + some_info['Bronze']

some_info.head()
```

# 3.5:Which sport exhibits the most balanced distribution across gold, silver, and bronze medals, and what might this indicate about competitiveness within that discipline?

# 3.6 :Which city hosted the most diverse Olympic program (in terms of number of distinct events), and what could this indicate about its infrastructure or global sporting influence?

```python
city_year_df = df[['City', 'Year', 'Event']].drop_duplicates()
city_year_df
```

```python
events_per_year = city_year_df.groupby(['City','Year']).size().unstack(fill_value=0)

plt.figure(figsize=(16,5))
events_per_year.plot(kind='bar', ax=plt.gca())
plt.title('Number of Events per City by Year of Summer Olympic medals \n 1976-to-2008')
plt.xlabel('City')
plt.ylabel('Number of Events held')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```

# 3.7: Which countries consistently appear in the top 10 across multiple Olympic Games from 1976 to 2008, and what factors might contribute to their sustained dominance in medal counts?

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(8,9))  # Adjust grid based on number of years
axs = axs.flatten()  # Flatten to 1D array for easy indexing
cols = [1976, 1980, 1984, 1988, 1992, 1996, 2000, 2004, 2008]

for idx, year in enumerate(cols):
    groupy = df[df['Year'] == year].groupby('Country')["Medal"].size().sort_values(ascending=False).head(10)
    axs[idx].bar(groupy.index, groupy.values)
    axs[idx].set_title(f'Top 10 Countries in {year}')
    axs[idx].tick_params(axis='x', rotation=90)
    axs[idx].grid(axis='y')



fig.suptitle(" Top 10 Countries by Medal Count Over the Years ", fontsize=18, fontweight='bold')
plt.tight_layout()
plt.show()
```
